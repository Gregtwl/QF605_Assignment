```python
import sympy as sp
import numpy as np
```

$$
IRR(S) = \sum_{i=1}^{(T_N - T_n \times m)}\frac{(1/m)}{(1+S/m)^i}
$$

```python
def compute_irr_derivatives():
    """
    Compute first and second derivatives of IRR(S) with respect to S
    IRR(S) = Σ(i=1 to T) (1/m) / (1+S/m)^i
    """
    # Define symbols
    S, m, i, T = sp.symbols('S m i T', real=True, positive=True)
    
    # Define the IRR function as a summation
    # Using finite summation from i=1 to T
    irr_expr = sp.Sum(1/m / (1 + S/m)**i, (i, 1, T))
    
    # Evaluate the summation (closed form for geometric series)
    irr_closed = (1/S) * (1 - (m/(m+S))**T)
    irr_simplified = sp.simplify(irr_closed)
    # irr_simplified = irr_closed

    
    print("IRR(S) Expression:")
    print(f"Original: {irr_expr}")
    print(f"\nClosed Form: {irr_simplified}")
    
    # Compute first derivative
    first_derivative = sp.diff(irr_simplified, S)
    first_derivative_simplified = sp.simplify(first_derivative)
    
    print(f"\n\nFirst Derivative (dIRR/dS):")
    print(first_derivative_simplified)
    
    # Compute second derivative
    second_derivative = sp.diff(first_derivative_simplified, S)
    second_derivative_simplified = sp.simplify(second_derivative)
    
    print(f"\n\nSecond Derivative (d²IRR/dS²):")
    print(second_derivative_simplified)
    
    return irr_simplified, first_derivative_simplified, second_derivative_simplified

# Run the function
irr, irr_prime, irr_double_prime = compute_irr_derivatives()
```

## IRR Derivatives - LaTeX Representation

**Original Expression (Summation Form):**
$$
\text{IRR}(S) 
= \sum_{i=1}^{T} \frac{\left(\frac{S}{m} + 1\right)^{- i}}{m}
$$

**Closed Form:**
$$
\text{IRR}_{\text{closed}}(S) 
= \frac{m^{T} \left(S + m\right)^{- T}}{S} + \frac{1}{S}
$$

**First Derivative:**
$$\frac{d\text{IRR}}{dS} 
= \frac{T \left(S + m\right)^{- 2 T} \left(S m + m^{2}\right)^{T}}{S \left(S + m\right)} + \frac{m^{T} \left(S + m\right)^{- T}}{S^{2}} - \frac{1}{S^{2}}

$$

**Second Derivative:**
$$
\frac{d^2\text{IRR}}{dS^2} 
= \frac{T^{2} \left(S + m\right)^{- 2 T} \left(S m + m^{2}\right)^{T}}{S \left(S + m\right)^{2}} - \frac{T \left(S + m\right)^{- 2 T} \left(S m + m^{2}\right)^{T}}{S \left(S + m\right)^{2}} - \frac{T m^{T} \left(S + m\right)^{- T}}{S^{2} \left(S + m\right)} - \frac{T \left(S + m\right)^{- 2 T} \left(S m + m^{2}\right) \left(S m + m^{2}\right)^{T}}{S^{2} m \left(S + m\right)^{2}} - \frac{2 m^{T} \left(S + m\right)^{- T}}{S^{3}} + \frac{2}{S^{3}}
